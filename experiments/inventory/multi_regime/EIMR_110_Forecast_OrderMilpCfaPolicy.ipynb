## 🚧 Problem definition

In this experiment we use our benchmark inventory problem with multi-regime Poisson demand and our standard set of cost parameters. Keeping this problem definition constant across experiments enables us to compare the effects of different policies while allowing demand intensity to vary with season, day, and weather.

```python
# === Defining the dynamic system ===

# Simulation horizon
T = 360

# Initial Inventory
# Start in (inventory=300, season=Summer=2, day=Monday=0, weather=Sunny=2)
S0= np.array([300.0, 2.0, 0.0, 2.0], dtype=float)

# Constraints
x_max = 480
dx = 10
S_max = 480

# Exogenous demand
exo = ExogenousPoissonMultiRegime(season_index=1, day_index=2, weather_index=3, season_period=90)

# Cost parameters 
p, c, h, b, K = 2.0, 0.5, 0.03, 1.0, 10.0
transition_func = lambda s, a, w, t: inventory_transition_regime_capped(s, a, w, t, s_max=S_max)
# Transition function
#transition_func = inventory_transition

# Cost functions
#cost_func = lambda s, a, w, t: inventory_cost(s, a, w, t, p=p, c=c, h=h, b=b)
cost_func = lambda s, a, w, t: inventory_cost_extended(s, a, w, t, p=p, c=c, h=h, b=b, K=K)

# Dynamic System
system = DynamicSystem(
    transition_func=transition_func,
    cost_func=cost_func,
    exogenous_model=exo,
    sim_seed=42,
    d_s=4,
    d_x=1,
    d_w=4,
    demand_history_window=7, # number of past demand observations included in state for forecasting, change to 7
 )

# Strict-CRN Monte Carlo evaluation settings
seed0 = 1234
n_episodes = 200
eval_info = make_eval_info(T=T, det_mode='argmax')
eval_info
```

## 🔮 Forecaster defintion



In this section, we specify which information model (a.k.a. **forecaster**) the policy uses to anticipate future demand in the multi-regime setting. We instantiate regime-aware forecaster variants, including a naive forecaster that repeats the last observed demand, a constant-mean forecaster that freezes the current regime-implied demand mean, and a learned ML forecaster trained on synthetic data from the same exogenous model.

Next, we conclude with a short sanity check to ensure that the resulting predictions are reasonable before they are embedded into the policy.

```python
# === Constant Forecaster ===
# Constant-mean forecaster: freeze the current regime-implied demand mean across the horizon.
forecaster_const = ConstantMeanForecaster(mean=float(350.0))

# Print example prediction: forecast mean demand path
H_demo = 5
mu_demo_const = forecaster_const.forecast_mean_path(S0, t=0, H=H_demo)
print(f"mu_demo_const (H={H_demo}):", np.round(mu_demo_const, 2))
```

```python
# === Naive Forecaster ===
# Naive forecaster: repeat the last observed demand for the next H steps.
# For the initial fallback, use the effective Poisson mean implied by the current multi-regime state.
mu0_multi_regime = exo.lambda_for_regimes(
    int(S0[exo.season_index]),
    int(S0[exo.day_index]),
    int(S0[exo.weather_index]),
)
forecaster_naive = NaiveForecaster(default_value=float(mu0_multi_regime))

# Print example prediction: Planned info payload passed into the CFA policy / forecaster call
naive_info_demo = {"last_demand": [295.0]}
H_demo = 5
mu_demo_naive = forecaster_naive.forecast_mean_path(S0, t=0, H=H_demo, info=naive_info_demo)
print("mu0_multi_regime:", float(mu0_multi_regime))
print("naive_info_demo:", naive_info_demo)
print(f"mu_demo_naive (H={H_demo}):", np.round(mu_demo_naive, 2))
```

```python
# === Rolling Mean Forecaster ===

# Rolling-mean forecaster: predict the mean of the last n observed demand values as next demand.
# During rollout, the simulator now provides a bounded demand_history when demand_history_window > 0.
forecaster_roll = RollingMeanForecaster(window_size=7, default_value=float(mu0_multi_regime))

# Print example prediction: forecast mean demand path
rolling_info_demo = {"demand_history": [300.0, 340.0, 365.0, 355.0]}
mu_demo_roll = forecaster_roll.forecast_mean_path(S0, t=0, H=H_demo, info=rolling_info_demo)
print("rolling history:", rolling_info_demo["demand_history"])
print("rolling-mean forecast:", mu_demo_roll)
```

```python
# === ML Forecaster ===
# Use the regime-aware adapter and forecaster so the model can condition on season, day, and weather.
adapter = MultiRegimeFeatureAdapter(exo)
forecaster_ml_tree = MlRegimeDemandForecaster(adapter, model_type="tree", random_state=0)

# Train on synthetic data generated from the multi-regime exogenous model
fit_seed = 7
val_seed = 99

forecaster_ml_tree.fit_from_exogenous(
    n_samples=6000,
    seed=fit_seed,
    t_start=0,
    val_samples=1000,
    val_seed=val_seed,
    val_t_start=8000,
 )

# Print ML forecaster fit report summary
print("ML forecaster fit_report (summary):")
print("  model_type:", forecaster_ml_tree.fit_report.get("model_type"))
print("  feature_dim:", forecaster_ml_tree.fit_report.get("feature_dim"))
for split in ("train", "val"):
    rep = forecaster_ml_tree.fit_report.get(split, {})
    mae = rep.get("MAE")
    rmse = rep.get("RMSE")
    bias = rep.get("Bias")
    n = rep.get("n_samples")
    print(f"  {split}: n={n} | MAE={mae:.3f} | RMSE={rmse:.3f} | Bias={bias:.3f}")

# Print example prediction: forecast mean demand path for the current multi-regime state
H_demo = 5
mu_demo = forecaster_ml_tree.forecast_mean_path(S0, t=0, H=H_demo)
print(f"\nmu_demo (H={H_demo}):", np.round(mu_demo, 2))
```

```python
# === ML AR(p) Forecaster — multi-regime ===
# Multi-regime AR(p): regime-transition features + last p realized demands.
# With p=7 and demand_history_window=7, the model uses D_{t-7}..D_{t-1} alongside
# the current season/day/weather transition probabilities.
#
# Model type choice and hyperparameters:
#
#   model_type="tree"       – DecisionTree.  tree_max_depth controls depth (default 6).
#                             With p=7 and n_samples=6000, max_depth=6 tends to overfit
#                             sparse lag combinations.  Reduce to 3 or 4 to regularise,
#                             or use "linear"/"elasticnet" instead.
#
#   model_type="linear"     – Ordinary least squares.  Each lag column gets one global
#                             weight; cannot memorize sparse leaves.  Generalises better
#                             when lags carry little independent signal beyond regimes.
#
#   model_type="elasticnet" – Regularised linear model.  alpha controls overall penalty
#                             strength; l1_ratio=1.0 is Lasso (sparse), l1_ratio=0.0 is
#                             Ridge (smooth shrinkage).  Best when you want uninformative
#                             lag weights shrunk toward zero automatically.
#
# Recommendation: "linear" or "elasticnet" for AR(p) with large p and moderate data.
# Use "tree" with tree_max_depth=3 as a compromise if you want non-linearity.

p_arp = 7  # ← set to match demand_history_window in DynamicSystem

adapter = MultiRegimeArpFeatureAdapter(exo, p=p_arp, lag_scale=100.0)
forecaster_ml_arp = MlArpDemandForecaster(
    adapter,
    model_type="linear",      # ← "linear" avoids tree overfitting on sparse lag combinations
    #model_type="elasticnet",  # ← regularised linear; tune alpha and l1_ratio below
    #model_type="tree",        # ← use with tree_max_depth=3 to reduce overfitting
    random_state=0,
    # --- tree hyperparameter (only used when model_type="tree") ---
    tree_max_depth=6,          # ← reduce to 3 or 4 to limit overfitting with large p
    # --- elasticnet hyperparameters (only used when model_type="elasticnet") ---
    alpha=1e-3,                # ← overall regularisation strength (larger = more shrinkage)
    l1_ratio=0.5,              # ← 0.0 = Ridge, 1.0 = Lasso, 0.5 = ElasticNet default
)

fit_seed = 7
val_seed  = 99

forecaster_ml_arp.fit_from_exogenous(
    n_samples=6000,
    seed=fit_seed,
    t_start=0,
    val_samples=1000,
    val_seed=val_seed,
    val_t_start=8000,
 )

# Print fit report
print("ML AR(p) (multi-regime) forecaster fit_report (summary):")
print("  model_type:", forecaster_ml_arp.fit_report.get("model_type"))
print("  feature_dim:", forecaster_ml_arp.fit_report.get("feature_dim"))
print("  p:", forecaster_ml_arp.fit_report.get("p"))
print("  tree_max_depth:", forecaster_ml_arp.fit_report.get("tree_max_depth"))
print("  alpha:", forecaster_ml_arp.fit_report.get("alpha"))
print("  l1_ratio:", forecaster_ml_arp.fit_report.get("l1_ratio"))
for split in ("train", "val"):
    rep = forecaster_ml_arp.fit_report.get(split, {})
    mae = rep.get("MAE")
    rmse = rep.get("RMSE")
    bias = rep.get("Bias")
    n = rep.get("n_samples")
    print(f"  {split}: n={n} | MAE={mae:.3f} | RMSE={rmse:.3f} | Bias={bias:.3f}")

# Sanity check: compare default forecast vs. forecast conditioned on recent demand history
H_demo = 5
mu_demo_default = forecaster_ml_arp.forecast_mean_path(S0, t=0, H=H_demo)
mu_demo_low = forecaster_ml_arp.forecast_mean_path(
    S0, t=0, H=H_demo,
    info={"demand_history": [150.0, 200.0, 180.0, 160.0, 170.0, 190.0, 155.0]},
)
mu_demo_high = forecaster_ml_arp.forecast_mean_path(
    S0, t=0, H=H_demo,
    info={"demand_history": [700.0, 750.0, 720.0, 710.0, 730.0, 760.0, 740.0]},
)
print(f"\nmu_demo_default     (H={H_demo}):", np.round(mu_demo_default, 2))
print(f"mu_demo_low_history  (H={H_demo}):", np.round(mu_demo_low, 2))
print(f"mu_demo_high_history (H={H_demo}):", np.round(mu_demo_high, 2))
```

## 🏋 Forecaster comparison

### Final five-forecaster comparison

To keep the comparison compact while still covering clearly different forecasting classes, we focus on five forecasters:

- `NaiveForecaster`: simple persistence baseline
- `SarimaxRegimeDemandForecaster`: classical statistical model with exogenous regime structure
- `MlRegimeDemandForecaster(model_type="tree")`: nonlinear regime-aware model without autoregression
- `MlAr1RegimeDemandForecaster(model_type="linear")`: linear history-aware model with one demand lag
